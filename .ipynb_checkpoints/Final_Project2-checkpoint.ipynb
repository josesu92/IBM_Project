```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import folium
import json
import urllib.request
from pandas.io.json import json_normalize 
import requests
```

Latitud y longitud de Chicago

```python
latitude = 41.881832
longitude = -87.623177
```

Lectura del archivo Json con los vecindarios

```python
url_json_file =r'https://raw.githubusercontent.com/blackmad/neighborhoods/master/chicago.geojson'
```

Se crea el mapa de los vecindarios que tiene Chicago

```python
f = folium.Figure(width=600, height=600)
chicago_map = folium.Map(location=[latitude, longitude], zoom_start=10).add_to(f)
folium.Choropleth(geo_data=url_json_file,fill_color='black',line_color='blue',fill_opacity=0.2,line_weight=1).add_to(chicago_map)
folium.Marker([latitude, longitude], popup='<i>Centre of Chicago</i>').add_to(chicago_map)
folium.Circle(radius=9000,location=[latitude,longitude],popup='The Waterfront',color='crimson',fill=True).add_to(chicago_map)
chicago_map
```
